# SIFT

```python
%matplotlib inline
import cv2
import numpy as np
from matplotlib import pyplot as plt

def SIFT_point(file_name):
    img = cv2.imread(file_name)
    gray= cv2.cvtColor(img,cv2.COLOR_BGR2GRAY)
    sift = cv2.SIFT()
    kp = sift.detect(gray,None)
    img=cv2.drawKeypoints(gray,kp)
    kk,des = sift.compute(gray,kp)
    """
    for a in kk:
        print a.size,a.angle
    """
    return img

a = SIFT_point('/Users/viplab/Desktop/1.jpg')
b = SIFT_point('/Users/viplab/Desktop/2.jpg')
c = SIFT_point('/Users/viplab/Desktop/3.jpg')

fig, (ax0,ax1,ax2) = plt.subplots(1,3, figsize=(13, 8))
ax0.imshow(a, cmap=plt.cm.gray, interpolation='nearest')
ax0.axis('off')
ax1.imshow(b, cmap=plt.cm.gray, interpolation='nearest')
ax1.axis('off')
ax2.imshow(c, cmap=plt.cm.gray, interpolation='nearest')
ax2.axis('off')

margins = dict(hspace=0.01, wspace=0.05, top=1, bottom=0, left=0, right=1)
fig.subplots_adjust(**margins)
```

# SURF

```python
def SURF_point(file_name):
    img = cv2.imread(file_name,0)

    # Create SURF object. You can specify params here or later.
    # Here I set Hessian Threshold to 400
    surf = cv2.SURF(400)
    surf.upright = True #don;t care orient
    # Find keypoints and descriptors directly
    kp, des = surf.detectAndCompute(img,None)
    if len(kp)>500:
        surf.hessianThreshold =25000
        kp, des = surf.detectAndCompute(img,None)
    elif len(kp)>300 and len(kp)<500:
        surf.hessianThreshold = 5000
        kp, des = surf.detectAndCompute(img,None)
    else:
        surf.hessianThreshold = 3000
        kp, des = surf.detectAndCompute(img,None)
        
    print "threshold:",surf.hessianThreshold

    img2 = cv2.drawKeypoints(img,kp,None,(255,0,0),4)
    print "kp",len(kp)
    print "size:",des.shape
    return img2

a = SURF_point('/Users/viplab/Desktop/1.jpg')
b = SURF_point('/Users/viplab/Desktop/2.jpg')
c = SURF_point('/Users/viplab/Desktop/3.jpg')

fig, (ax0,ax1,ax2) = plt.subplots(1,3, figsize=(13, 8))
ax0.imshow(a, cmap=plt.cm.gray, interpolation='nearest')
ax0.axis('off')
ax1.imshow(b, cmap=plt.cm.gray, interpolation='nearest')
ax1.axis('off')
ax2.imshow(c, cmap=plt.cm.gray, interpolation='nearest')
ax2.axis('off')

margins = dict(hspace=0.01, wspace=0.05, top=1, bottom=0, left=0, right=1)
fig.subplots_adjust(**margins)
```

# Brief

```python
def BRIEF_point(file_name):
    img = cv2.imread(file_name,0)

    # Initiate STAR detector
    star = cv2.FeatureDetector_create("STAR")

    # Initiate BRIEF extractor
    brief = cv2.DescriptorExtractor_create("BRIEF")

    # find the keypoints with STAR
    kp = star.detect(img,None)

    # compute the descriptors with BRIEF
    kp, des = brief.compute(img, kp)
    ig = cv2.drawKeypoints(img,kp,None,(255,0,0),4)
    print des.shape,len(kp)
    return ig

a = BRIEF_point('/Users/viplab/Desktop/1.jpg')
b = BRIEF_point('/Users/viplab/Desktop/2.jpg')
c = BRIEF_point('/Users/viplab/Desktop/3.jpg')

fig, (ax0,ax1,ax2) = plt.subplots(1,3, figsize=(13, 8))
ax0.imshow(a, cmap=plt.cm.gray, interpolation='nearest')
ax0.axis('off')
ax1.imshow(b, cmap=plt.cm.gray, interpolation='nearest')
ax1.axis('off')
ax2.imshow(c, cmap=plt.cm.gray, interpolation='nearest')
ax2.axis('off')

margins = dict(hspace=0.01, wspace=0.05, top=1, bottom=0, left=0, right=1)
fig.subplots_adjust(**margins)
```

# ORB

```python
def ORB_point(file_name):
    img = cv2.imread(file_name,0)
    # Initiate STAR detector
    orb = cv2.ORB()

    # find the keypoints with ORB
    kp = orb.detect(img,None)

    # compute the descriptors with ORB
    kp, des = orb.compute(img, kp)

    # draw only keypoints location,not size and orientation
    img2 = cv2.drawKeypoints(img,kp,color=(255,0,0), flags=0)
    return img2
    
a = ORB_point('/Users/viplab/Desktop/1.jpg')
b = ORB_point('/Users/viplab/Desktop/2.jpg')
c = ORB_point('/Users/viplab/Desktop/3.jpg')

fig, (ax0,ax1,ax2) = plt.subplots(1,3, figsize=(13, 8))
ax0.imshow(a, cmap=plt.cm.gray, interpolation='nearest')
ax0.axis('off')
ax1.imshow(b, cmap=plt.cm.gray, interpolation='nearest')
ax1.axis('off')
ax2.imshow(c, cmap=plt.cm.gray, interpolation='nearest')
ax2.axis('off')

margins = dict(hspace=0.01, wspace=0.05, top=1, bottom=0, left=0, right=1)
fig.subplots_adjust(**margins)
```
